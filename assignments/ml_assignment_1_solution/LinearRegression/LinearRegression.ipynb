### 1.5.2 Training

#### train_regression1.py

```python
# Given petal length and petal width, predict sepal length.

petal_iris_data = load_iris()

petal_df = pd.DataFrame(petal_iris_data.data, columns=petal_iris_data.feature_names)

X_lr_pl_pw_sl = petal_df[['petal length (cm)', 'petal width (cm)']].values
y_lr_pl_pw_sl = petal_df['sepal length (cm)'].values

split_index = int(0.9 * len(X_lr_pl_pw_sl))
X_lr_pl_pw_sl_train, X_lr_pl_pw_sl_test = X_lr_pl_pw_sl[:split_index], X_lr_pl_pw_sl[split_index:]
y_lr_pl_pw_sl_train, y_lr_pl_pw_sl_test = y_lr_pl_pw_sl[:split_index], y_lr_pl_pw_sl[split_index:]

lr_pl_pw_sl = LR()

# Load the model if it has been trained before
lr_pl_pw_sl.load('lr_pl_pw_sl.pkl')
if not lr_pl_pw_sl.loaded:
    lr_pl_pw_sl.fit(X_lr_pl_pw_sl_train, y_lr_pl_pw_sl_train)
    lr_pl_pw_sl.save('lr_pl_pw_sl.pkl')
pl_pw_sl_history = lr_pl_pw_sl.loss_history
```

### 1.5.3 Regularization

#### regularization_train.py

```python
# Given petal length and petal width, predict sepal length with regularization.

petal_iris_data = load_iris()

petal_df = pd.DataFrame(petal_iris_data.data, columns=petal_iris_data.feature_names)

X_regularization = petal_df[['sepal length (cm)', 'sepal width (cm)']].values
y_regularization = petal_df['petal length (cm)'].values

split_index = int(0.9 * len(X_regularization))
X_regularization_train, X_regularization_test = X_regularization[:split_index], X_regularization[split_index:]
y_regularization_train, y_regularization_test = y_regularization[:split_index], y_regularization[split_index:]

lr_regularization = LR()

# Load the model if it has been trained before
lr_regularization.load('lr_regularization.pkl')
if not lr_regularization.loaded:
    lr_regularization.fit(X_regularization_train, y_regularization_train, regularization=2.0)
    lr_regularization.save('lr_regularization.pkl')
regularization_history = lr_regularization.loss_history

# Plot the loss history
plt.plot(regularization_history, label='Regularization')
```

#### Comparing with and without regularization

```python
# Load the model without regularization
lr_pl_pw_sl.load('lr_sl_sw_pl.pkl')

# Load the model with regularization
lr_regularization.load('lr_regularization.pkl')

# Plot the loss history for both models
plt.plot(sl_sw_pl_history, label='Without Regularization', color='red')
plt.plot(regularization_history, label='With Regularization', color='green')

print('Weight of the model without regularization:', lr_sl_sw_pl.weights)
print('Bias of the model without regularization:', lr_sl_sw_pl.bias)
print('Weight of the model with regularization:', lr_regularization.weights)
print('Bias of the model with regularization:', lr_regularization.bias)
```
