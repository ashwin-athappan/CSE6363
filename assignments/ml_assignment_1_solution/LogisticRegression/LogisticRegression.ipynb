```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
```

```python
# Import LinearRegression class from LinearRegression.py

import importlib
import LogisticRegression

importlib.reload(LogisticRegression)

lgr = LogisticRegression.LogisticRegression
```

## 2.1 Logistic Regression

#### train_classification1.py

```python
# Classification with sepal length and sepal width
# Input: Sepal Length, Sepal Width
# Output: Iris Species

iris_data = load_iris()

petal_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
petal_df['target'] = iris_data.target
X_lgr_sl_sw = petal_df[['sepal length (cm)', 'sepal width (cm)']].values
y_lgr_sl_sw = petal_df['target'].values

X_lgr_sl_sw_train, X_lgr_sl_sw_test, y_lgr_sl_sw_train, y_lgr_sl_sw_test = train_test_split(
    X_lgr_sl_sw, y_lgr_sl_sw, test_size=0.1, random_state=42)

lgr_sl_sw = lgr()
lgr_sl_sw.fit(X_lgr_sl_sw_train, y_lgr_sl_sw_train)
plot_decision_regions(X_lgr_sl_sw_train, y_lgr_sl_sw_train, lgr_sl_sw)
plt.title("Logistic Regression")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
# Accuracy for dual-feature (Sepal Length, Sepal Width) logistic regression

val_logits = np.dot(X_lgr_sl_sw_train, lgr_sl_sw.weights) + lgr_sl_sw.bias
val_probs = lgr_sl_sw.softmax(val_logits)
accuracy = np.mean(np.argmax(val_probs, axis=1) == y_lgr_sl_sw_train)
accuracy * 100
```

#### train_classification2.py

#### train_classification3.py

```python
# Classification with sepal length, sepal width, petal length, and petal width
# Input: Petal Length, Petal Width, Sepal Length, Sepal Width
# Output: Iris Species

iris_data = load_iris()

petal_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
petal_df['target'] = iris_data.target
X_lgr_sl_sw_pl_pw = petal_df[['petal length (cm)', 'petal width (cm)', 'sepal length (cm)', 'sepal width (cm)']].values
y_lgr_sl_sw_pl_pw = petal_df['target'].values

X_lgr_sl_sw_pl_pw_train, X_lgr_sl_sw_pl_pw_test, y_lgr_sl_sw_pl_pw_train, y_lgr_sl_sw_pl_pw_test = train_test_split(
    X_lgr_sl_sw_pl_pw, y_lgr_sl_sw_pl_pw, test_size=0.1, random_state=42)

lgr_sl_sw_pl_pw = lgr()
lgr_sl_sw_pl_pw.fit(X_lgr_sl_sw_pl_pw_train, y_lgr_sl_sw_pl_pw_train)
plot_decision_regions(X_lgr_sl_sw_pl_pw_train, y_lgr_sl_sw_pl_pw_train, lgr_sl_sw_pl_pw, filler_feature_values={2: 5.0, 3: 3.5}, filler_feature_ranges={2: 2.0, 3: 1.0})
plt.title("Logistic Regression")
plt.xlabel("Petal Length, Petal Width")
plt.ylabel("Sepal Length, Sepal Width")
plt.show()
```

```python
# Accuracy for all-feature logistic regression

logits = np.dot(X_lgr_sl_sw_pl_pw_train, lgr_sl_sw_pl_pw.weights) + lgr_sl_sw_pl_pw.bias
probs = lgr_sl_sw_pl_pw.softmax(logits)
accuracy = np.mean(np.argmax(probs, axis=1) == y_lgr_sl_sw_pl_pw_train)
accuracy * 100
```
