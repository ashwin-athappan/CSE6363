```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from mlxtend.plotting import plot_decision_regions
```

```python
# Import LinearRegression class from LinearRegression.py

import importlib
import LogisticRegression

importlib.reload(LogisticRegression)

lgr = LogisticRegression.LogisticRegression
```

```python
iris_data = load_iris()

petal_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
petal_df['target'] = iris_data.target
X_lgr_sl_sw_pl_pw = petal_df[['sepal length (cm)', 'sepal width (cm)']].values
y_lgr_sl_sw_pl_pw = petal_df['target'].values

X_lgr_sl_sw_pl_pw_train, X_lgr_sl_sw_pl_pw_test, y_lgr_sl_sw_pl_pw_train, y_lgr_sl_sw_pl_pw_test = train_test_split(X_lgr_sl_sw_pl_pw, y_lgr_sl_sw_pl_pw, test_size=0.1, random_state=42)


lgr_sl_sw_pl_pw = lgr()
lgr_sl_sw_pl_pw.fit(X_lgr_sl_sw_pl_pw_train, y_lgr_sl_sw_pl_pw_train)
plot_decision_regions(X_lgr_sl_sw_pl_pw_train, y_lgr_sl_sw_pl_pw_train, lgr_sl_sw_pl_pw)
plt.title("Logistic Regression")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

```python
plt.plot(range(len(lgr_sl_sw_pl_pw.loss_history)), lgr_sl_sw_pl_pw.loss_history)
```
